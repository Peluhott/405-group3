```python
import pandas as pd
# Load the data from the excel file
df = pd.read_excel('data.xlsx')
```

```python
# Display the different columns in the dataset
print(df.columns)
print()
print(df.head())
print(df.shape)

missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)


```

The dataset has 5000 records with 6 columns. It's currently only missing values in certifications, which represents that the person does not have any certifications.

```python
print(df.nunique())
print()
print(df["Education Level"].unique())
print()
print(df["Specialization"].unique())
print()
print(df["Certifications"].unique())
print()

print(df["Recommended Career"].unique())
```

```python
unique_skills = (
    df["Skills"]
      .str.split(",")
      .explode()
      .str.strip()
      .unique()
)

print(len(unique_skills))
print(unique_skills)
```

All columns are standardized , ready to be used. Skills has 10 unique values, the reason it showed 757 unique values is because its counting different combinations of the 10 values as unique.
